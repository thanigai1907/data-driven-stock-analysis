```python
import pandas as pd

# Load CSV
df = pd.read_csv(r'C:\Users\velut\OneDrive\Documents\New folder\output\all_stocks.csv')

# View Data
print(df.head())
```

```python
df['date'] = pd.to_datetime(df['date'])
df['MonthYear'] = df['date'].dt.strftime('%b-%Y')

print(df.head())
```

```python
df['Daily_Return'] = ((df['close'] - df['open']) / df['open']) * 100

print(df.head())
```

```python
#Step 3: Group by MonthYear and Ticker, then sum the Daily_Return
monthly_returns = df.groupby(['MonthYear', 'Ticker'])['Daily_Return'].sum().reset_index()

print(monthly_returns.head())

# Step 4: Save to CSV
output_path = r'C:\Users\velut\OneDrive\Documents\New folder\output\monthly_returns.csv'
monthly_returns.to_csv(output_path, index=False)

print(f'File saved successfully to {output_path}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the monthly returns data
df = pd.read_csv( r'C:\Users\velut\OneDrive\Documents\New folder\output\monthly_returns.csv')

# Get unique months
months = df['MonthYear'].unique()

# Set Seaborn style
sns.set(style="whitegrid")

# Loop through each month
for month in months:
    # Filter data for that month
    month_df = df[df['MonthYear'] == month]

    # Top 5 Gainers
    top_gainers = month_df.nlargest(5, 'Daily_Return')

    # Top 5 Losers
    top_losers = month_df.nsmallest(5, 'Daily_Return')

    # Plotting
    plt.figure(figsize=(12, 6))
    
    # Top 5 Gainers
    plt.subplot(1, 2, 1)
    sns.barplot(x='Daily_Return', y='Ticker', data=top_gainers, palette='Greens')
    plt.title(f'{month} - Top 5 Gainers')

    # Top 5 Losers
    plt.subplot(1, 2, 2)
    sns.barplot(x='Daily_Return', y='Ticker', data=top_losers, palette='Reds')
    plt.title(f'{month} - Top 5 Losers')

    plt.suptitle(f'Stock Performance for {month}', fontsize=16)
    plt.tight_layout()
    
    # Show plot
    plt.show()
```
